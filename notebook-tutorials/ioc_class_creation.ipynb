# IOC Class Creation
## Introduction
For **most** ibek users an image containing all binaries, dbd files and an ibek readable `<ioc_class>.ibek.yaml` file will already exist, and the job of a developer is simply to use that image to create an instance as shown [here](creating_ioc_instances.ipynb).

In this tutorial we will look at how developers can add ibek support to new IOC classes for containerised deployment. This tutorial assumes you already have a built image containing all IOC binaries and dbd files needed for this IOC class.

For this to work you will need:
- ibek
- An `ibek.schema.json` schema
- An `<ioc_class>.ibek.yaml` file

## Generating `ibek.schema.json`
This schema is generic for all classes of IOC and is generated using the `ibek-schema` function in ibek. An example of this is shown below:

```bash
ibek=../ibek/
ibek_schema=/tmp/ibek.schema.json
python $ibek ibek-schema $ibek_schema
```

```bash
cat $ibek_schema
```

## Constructing `<ioc_class>.ibek.yaml`
Each class of IOC is defined by an `<ioc_class>.ibek.yaml` file. This file is required by ibek when [creating an IOC instance](creating_ioc_instances.ipynb). `<ioc_class>.ibek.yaml` must adhere to the schema `ibek.schema.json`. To ensure that it does, we recommend using a schema aware editor to assist you.
